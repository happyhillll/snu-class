## 필요한 라이브러리 불러오기


```python
import requests
import pandas as pd
```

## 1. 발의 법률안 데이터 수집하기


```python
url = "https://open.assembly.go.kr/portal/openapi/nzmimeepazxkubdpn?KEY=640b6edf35bb47f280aac61745395c9e&AGE=21&Type=json"
max_page = requests.get(url).json()["nzmimeepazxkubdpn"][0]["head"][0][
    "list_total_count"
]
laws_21 = []
for page_num in range(1, int(max_page / 1000 + 2)):
    p_size = page_num * 1000
    p_index = page_num
    url = f"https://open.assembly.go.kr/portal/openapi/nzmimeepazxkubdpn?KEY=640b6edf35bb47f280aac61745395c9e&AGE=21&Type=json&pIndex={page_num}&pSize=1000"
    laws = requests.get(url).json()["nzmimeepazxkubdpn"][1]["row"]
    laws_21.extend(laws)
laws_21 = pd.DataFrame(laws_21)
```

```python
# 데이터의 앞 부분 10개만 출력
laws_21.head(10)
```

```python
# 간단한 통계 정보 출력
laws_21.describe()
```

```python
# 사라질 수도 있으므로 data를 저장해둡니다.
laws_21.to_csv("./data/laws_21.csv", index=False)
```

```python
# 20대 국회의원 발의 법률안도 수집하고 저장합니다.
url = "https://open.assembly.go.kr/portal/openapi/nzmimeepazxkubdpn?KEY=640b6edf35bb47f280aac61745395c9e&AGE=20&Type=json"
max_page = requests.get(url).json()["nzmimeepazxkubdpn"][0]["head"][0][
    "list_total_count"
]
laws_20 = []
for page_num in range(1, int(max_page / 1000 + 2)):
    p_size = page_num * 1000
    p_index = page_num
    url = f"https://open.assembly.go.kr/portal/openapi/nzmimeepazxkubdpn?KEY=640b6edf35bb47f280aac61745395c9e&AGE=20&Type=json&pIndex={page_num}&pSize=1000"
    laws = requests.get(url).json()["nzmimeepazxkubdpn"][1]["row"]
    laws_20.extend(laws)
laws_20 = pd.DataFrame(laws_20)
laws_20.to_csv("./data/laws_20.csv", index=False)
```

## 2. 국회의원 인적사항 데이터 수집하기


```python
url = "https://open.assembly.go.kr/portal/openapi/npffdutiapkzbfyvr?KEY=640b6edf35bb47f280aac61745395c9e&Type=json&UNIT_CD=100020&pSize=1000"
lawmaker_20 = requests.get(url).json()["npffdutiapkzbfyvr"][1]["row"]
lawmaker_20 = pd.DataFrame(lawmaker_20)
lawmaker_20.to_csv("./data/lawmaker_20.csv", index=False)
```

```python
url = "https://open.assembly.go.kr/portal/openapi/nwvrqwxyaytdsfvhu?KEY=640b6edf35bb47f280aac61745395c9e&Type=json&pSize=1000"
lawmaker_21 = requests.get(url).json()["nwvrqwxyaytdsfvhu"][1]["row"]
lawmaker_21 = pd.DataFrame(lawmaker_21)
# 제21대 국회의원이었다가 제외된 국회의원들도 목록에 추가해 줍니다.
url = "https://open.assembly.go.kr/portal/openapi/npffdutiapkzbfyvr?KEY=640b6edf35bb47f280aac61745395c9e&Type=json&UNIT_CD=100021&pSize=1000"
lawmaker_21_missing = requests.get(url).json()["npffdutiapkzbfyvr"][1]["row"]
lawmaker_21_missing = pd.DataFrame(lawmaker_21_missing)
lawmaker_21 = pd.concat([lawmaker_21, lawmaker_21_missing])
lawmaker_21.to_csv("./data/lawmaker_21.csv", index=False)
```
